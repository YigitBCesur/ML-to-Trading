**Özetle**

* **Portföy teorisinin bileşenleri:**
    * **Ağırlıklar (1'e eşit olmalı):** Bu, portföyünüzdeki her bir varlığa yatırılan toplam yatırımınızın oranını ifade eder. Örneğin, A, B ve C hisselerinden oluşan bir portföyünüz varsa, ağırlıkları (w_A, w_B, w_C) her bir hisseye yatırılan paranın yüzdesini temsil eder. Ağırlıklar 1'e (veya %100'e) eşit olmalıdır, çünkü yatırımınızın tamamını tahsis ediyorsunuz.
    * **Getiri, Ortalama, Kovaryans:**
        * **Getiri:** Bir yatırımın belirli bir süre içindeki gerçek kazancı veya kaybı.
        * **Ortalama (Beklenen Getiri):** Geçmiş verilere ve/veya tahminlere dayanarak bir yatırımdan elde etmeyi umduğunuz ortalama getiri.
        * **Kovaryans:** İki varlığın getirilerinin birbirleriyle nasıl hareket ettiğinin istatistiksel bir ölçüsü.  Çeşitlendirmenin riski nasıl azaltabileceğini anlamaya yardımcı olur.

* **2 Varlık Durumu (Ortalama, Varyans, Parabol):** Bu, yalnızca iki varlıktan oluşan basitleştirilmiş bir portföyü ifade eder. Portföy ortalaması (beklenen getiri) ve varyansı (risk) kombinasyonlarını çizdiğinizde, genellikle parabolik bir eğri elde edersiniz. Bu eğri, iki varlıklı bir portföy için **etkin sınırı** temsil eder ve en iyi olası risk-getiri dengelerini gösterir.

* **D Varlık Durumu için Ortalama ve Varyans:** Bu, kavramları 'D' sayıda varlığı olan bir portföye (2'den büyük herhangi bir sayı) genelleştirir. Portföy ortalamasını ve varyansını hesaplamak, tüm varlık çiftleri arasındaki kovaryansı hesaba katmanız gerektiğinden daha karmaşık hale gelir.

* **Ortalama-varyans optimizasyonu, Sharpe oranı vb. için kullanılacaktır:** Bu, tartışılan kavramların daha gelişmiş portföy analiz teknikleri için yapı taşları olduğunu gösterir:
    * **Ortalama-Varyans Optimizasyonu:** Belirli bir risk seviyesi için beklenen getiriyi en üst düzeye çıkaran (veya belirli bir beklenen getiri için riski en aza indiren) portföy ağırlıklarını bulma süreci.
    * **Sharpe Oranı:** Aldığınız her birim oynaklık için ne kadar fazla getiri elde ettiğinizi gösteren, riskten arındırılmış getiri ölçüsü.

**Portföy Getirisinin Ortalama-Varyansı Açıklaması**

Temel fikir, bir portföy oluştururken hem beklenen getiriyi hem de o portföyün riskini (varyans veya standart sapma) göz önünde bulundurmanız gerektiğidir.

**1. Portföy Beklenen Getirisi:**

Bir portföyün beklenen getirisi, basitçe portföydeki bireysel varlıkların beklenen getirilerinin ağırlıklı ortalamasıdır.

**Formül:** E(R_p) = Σ (w_i * E(R_i))
   * Burada:
      * E(R_p) portföyün beklenen getirisidir.
      * w_i, portföydeki i varlığının ağırlığıdır.
      * E(R_i), i varlığının beklenen getirisidir.

**2. Portföy Varyansı (Riski):**

Bir portföyün varyansı, bireysel varlık varyanslarının ağırlıklı ortalaması DEĞİLDİR. Bunun nedeni, çeşitlendirmenin riski azaltabilmesidir. Buradaki anahtar, varlıklar arasındaki kovaryansı dikkate almaktır.

**Formül (2 varlıklı bir portföy için):**

Var(R_p) = w_1^2 * Var(R_1) + w_2^2 * Var(R_2) + 2 * w_1 * w_2 * Cov(R_1, R_2)

   * Burada:
      * Var(R_p) portföyün varyansıdır.
      * w_1 ve w_2, sırasıyla 1 ve 2 numaralı varlıkların ağırlıklarıdır.
      * Var(R_1) ve Var(R_2), sırasıyla 1 ve 2 numaralı varlıkların varyanslarıdır.
      * Cov(R_1, R_2), 1 ve 2 numaralı varlıklar arasındaki kovaryanstır.

**Temel Çıkarımlar:**

* **Çeşitlendirme Önemlidir:** Varyans formülündeki kovaryans terimi çok önemlidir. Mükemmel bir şekilde ilişkili olmayan (yani, düşük veya negatif kovaryansa sahip) varlıkları birleştirerek, beklenen getiriden ödün vermeden genel portföy varyansını (riski) azaltabilirsiniz.
* **Etkin Sınır:** Belirli bir risk seviyesi için en yüksek beklenen getiriyi (veya belirli bir beklenen getiri için en düşük riski) sunan portföyler kümesine etkin sınır denir. Yatırımcılar bu sınırda yer alan portföylere sahip olmayı hedeflerler.
* **Denge:** Risk ve getiri arasında doğal bir denge vardır. Daha yüksek beklenen getiriler genellikle daha yüksek riskle gelir ve bunun tersi de geçerlidir. Portföy teorisi, yatırımcıların risk toleranslarına göre en uygun dengeyi bulmalarına yardımcı olur.

**Özetle, bu slayt, bir portföy bağlamında risk ve getiriyi nasıl ölçüleceğini ve yönetileceğini anlamak için bir yol haritası sunar. Yatırım yönetimiyle uğraşan herkes için temel bir kavramdır.**

###**---2-Asset-Portfolio Optimization----**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(11111111)
```

```python
mean_return = 0.01 * np.random.randn(2)
mean_return
```

```python
rho = 0.01 * np.random.randn()
rho
```

```python
sigmas = np.exp(np.random.randn(2))
sigmas
```

```python
cov = np.diag(sigmas**2)
cov
```

```python
sigma12 = sigmas[0] * sigmas[1] * rho
cov[0,1] = sigma12
cov[1,0] = sigma12
cov
```

```python
def softmax(w):
  a = np.exp(w)
  return a / a.sum()
```

```python
N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
  w = softmax(np.random.randn(2))
  ret = mean_return.dot(w)
  risk = np.sqrt(w.dot(cov).dot(w))
  returns[i] = ret
  risks[i] = risk
```

```python
plt.scatter(risks, returns, alpha=0.1);
plt.xlabel("Risk");
plt.ylabel("Return");
```

```python
N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
  x = np.random.random()
  w = np.array([x, 1-x]) # positive-only
  ret = mean_return.dot(w)
  risk = np.sqrt(w.dot(cov).dot(w))
  returns[i] = ret
  risks[i] = risk
```

```python
plt.scatter(risks, returns, alpha=0.1);
plt.xlabel("Risk");
plt.ylabel("Return");
```

```python
N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
  x = np.random.random() - 0.5
  w = np.array([x, 1-x]) # short-selling allowed
  ret = mean_return.dot(w)
  risk = np.sqrt(w.dot(cov).dot(w))
  returns[i] = ret
  risks[i] = risk
```

```python
plt.scatter(risks, returns, alpha=0.1);
```

```python
G = matrix(0.0, (3,3))
G[::4] = -1.0
print(G)
```

```python
def my_minimize(mu, cov, target_return):
  mu2 = matrix(mu)
  cov2 = matrix(2*cov)
  q = matrix(0.0, (D, 1))
  # Gx <= h

  # makes identity matrix
  # G = matrix(0.0, (D, D))
  # G[::D+1] = -1.0
  G = matrix(-np.eye(D))
  h = matrix(0.5, (D, 1))
  # equivalent to
  # w(i) >= -h(i)

  # Ax = b
  # A = matrix(1.0, (1, D))
  A = np.vstack((
      np.ones(D),
      mu,
  ))
  A = matrix(A)
  b = matrix([1.0, target_return])

  res = qp(cov2, q, G, h, A, b)
  if res['status'] != 'optimal':
    print(res)
  return np.sqrt(res['primal objective']), np.array(res['x'])
```

###**Sharpe Ratio**

[Buradan Bakiniz :   3-Month Treasury Bill Secondary Market Rate, Discount Basis (TB3MS)](https://fred.stlouisfed.org/series/TB3MS)

```python
risk_free_rate = 0.03 / 252
```

```python
def neg_sharpe_ratio(weights):
  mean = weights.dot(mean_return)
  sd = np.sqrt(weights.dot(cov).dot(weights))
  return -(mean - risk_free_rate) / sd
```

```python
res = minimize(
    fun=neg_sharpe_ratio,
    x0=np.ones(D) / D, # uniform
    method='SLSQP',
    constraints={
        'type': 'eq',
        'fun': portfolio_constraint,
    },
    bounds=bounds,
)
res
```

```python
best_sr, best_w = -res.fun, res.x
```

```python
mc_best_w = None
mc_best_sr = float('-inf')
for i, (risk, ret) in enumerate(zip(risks, returns)):
  sr = (ret - risk_free_rate) / risk
  if sr > mc_best_sr:
    mc_best_sr = sr
    mc_best_w = random_weights[i]
print(mc_best_w, mc_best_sr)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
plt.scatter(risks, returns, alpha=0.1);
plt.plot(optimized_risks, target_returns, c='black');

# found by optimization
opt_risk = np.sqrt(best_w.dot(cov).dot(best_w))
opt_ret = mean_return.dot(best_w)
plt.scatter([opt_risk], [opt_ret], c='red');

# found by monte carlo simulation
mc_risk = np.sqrt(mc_best_w.dot(cov).dot(mc_best_w))
mc_ret = mean_return.dot(mc_best_w)
plt.scatter([mc_risk], [mc_ret], c='orange');
```
